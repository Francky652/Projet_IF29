# Étape 4 : Expérimentation et analyse comparative

## 4.1 Objectif de la comparaison

Après avoir mis en œuvre une approche non supervisée (K-Means) et une approche supervisée (Random Forest), l'objectif est maintenant de comparer les résultats obtenus afin d'évaluer leur capacité à identifier des profils atypiques parmi les utilisateurs Twitter.

Dans notre étude, nous considérons comme atypique un utilisateur dont le comportement s'écarte significativement de celui observé chez la majorité des comptes. Cette atypicité peut se traduire par une activité inhabituelle, une visibilité très faible malgré une forte activité, ou encore des comportements proches de ceux observés chez certains comptes automatisés. Cette définition repose uniquement sur les indicateurs disponibles dans le dataset et ne permet pas d'affirmer qu'un compte est effectivement un bot ou un spammeur.


## 4.2 Hypothèses de travail

Avant d'analyser les résultats, nous avons formulé plusieurs hypothèses.

### H1 : Les profils atypiques se regroupent naturellement dans certains clusters.

L'idée est que des utilisateurs ayant des comportements similaires devraient être rapprochés par K-Means, ce qui pourrait faire apparaître des groupes de comptes suspects ou fortement automatisés.

### H2 : Les caractéristiques liées à la popularité et à l'activité des comptes jouent un rôle important dans l'identification des profils atypiques.

Nous supposons que des variables comme le nombre de followers, le nombre d'abonnements ou l'activité globale du compte sont de bons indicateurs pour distinguer les utilisateurs atypiques du reste de la population.

### H3 : L'approche supervisée permet une détection plus précise des profils atypiques.

Une fois une définition métier du profil atypique établie, nous nous attendons à ce qu'un modèle supervisé soit plus performant qu'une approche de clustering pour identifier automatiquement ces profils.


## 4.3 Résultats obtenus avec K-Means

Le clustering a permis de mettre en évidence plusieurs comportements distincts parmi les utilisateurs.

Avec la PCA, nous avons principalement observé :

* un groupe très important d'amplificateurs passifs qui relaient massivement le contenu des autres utilisateurs ;
* un groupe de créateurs de contenu plus actifs ;
* quelques profils extrêmement visibles correspondant à des célébrités ou à de grands médias ;
* un petit groupe de super-diffuseurs produisant une quantité importante de contenu.

Sans PCA, la segmentation devient plus fine. Nous avons notamment réussi à isoler un cluster composé de comptes partageant presque systématiquement des liens externes et réalisant un très grand nombre de citations de tweets. Ce comportement est particulièrement atypique par rapport au reste de la population et peut correspondre à des comptes automatisés ou à des agrégateurs de contenu.

Le principal avantage de K-Means est qu'il permet de découvrir ces profils sans avoir besoin de définir à l'avance ce qu'est un compte atypique.


## 4.4 Résultats obtenus avec Random Forest

Pour l'approche supervisée, nous avons construit une variable cible à partir de plusieurs critères comportementaux :

* ratio followers/friends très faible ;
* activité anormalement élevée ;
* utilisation importante de hashtags ;
* faible visibilité.

Un utilisateur satisfaisant au moins deux de ces critères est considéré comme atypique.

Cette définition conduit à identifier 206 024 profils atypiques parmi les 1 843 439 utilisateurs du dataset, soit environ 11,18 % de la population étudiée.

Les performances obtenues sont les suivantes :

* Accuracy : 97 %
* Precision (classe atypique) : 80 %
* Recall (classe atypique) : 100 %
* F1-score : 88 %

Le modèle détecte donc pratiquement tous les profils atypiques définis par nos critères, avec très peu de faux négatifs.


## 4.5 Analyse des variables importantes

L'étude de l'importance des variables montre que les attributs les plus influents sont :

1. followers_count ;
2. listed_count ;
3. statuses_count ;
4. friends_count.

Le nombre de followers représente à lui seul près de 64 % de l'importance totale du modèle.

Cela suggère que la popularité et la visibilité d'un compte jouent un rôle majeur dans la distinction entre profils normaux et profils atypiques.

À l'inverse, certaines variables comme le nombre de réponses ou le nombre de citations ont eu un impact beaucoup plus faible que prévu.


## 4.6 Comparaison des deux approches

Les deux méthodes ne répondent pas exactement au même objectif.

Le K-Means cherche avant tout à explorer les données et à faire apparaître des groupes d'utilisateurs présentant des comportements similaires. Il permet ainsi de découvrir des profils inattendus ou de mettre en évidence des comportements particuliers.

Le Random Forest cherche au contraire à reproduire une définition précise du profil atypique. Une fois entraîné, il est capable de classer automatiquement les nouveaux utilisateurs.

Le clustering apporte donc une meilleure compréhension de la structure globale des données, tandis que le modèle supervisé apporte une meilleure capacité de décision.


## 4.7 Validation des hypothèses

### Hypothèse H1

Les profils atypiques se regroupent naturellement dans certains clusters.

Cette hypothèse semble validée.

Le K-Means a permis d'isoler plusieurs groupes présentant des comportements très différents du reste des utilisateurs, notamment les comptes partageant massivement des liens externes ou relayant automatiquement du contenu.

### Hypothèse H2

Les variables liées à la popularité et à l'activité jouent un rôle important dans la détection des profils atypiques.

Cette hypothèse est également validée.

Les variables les plus importantes du Random Forest sont directement liées à la visibilité et à l'activité des comptes, notamment le nombre de followers, le nombre d'apparitions dans des listes et le volume d'activité historique.

### Hypothèse H3

L'approche supervisée permet une détection plus précise des profils atypiques.

Les résultats obtenus vont dans ce sens.

Le Random Forest atteint un F1-score de 88 % et un rappel de 100 %, ce qui montre qu'il est capable de retrouver efficacement les profils atypiques définis dans notre étude.


## 4.8 Limites et pistes d'amélioration

La principale limite de ce travail est l'absence de labels réels indiquant quels comptes sont effectivement des bots ou des spammeurs.

Notre variable cible repose sur une définition métier construite à partir des données disponibles. Les profils détectés doivent donc être considérés comme potentiellement atypiques plutôt que comme des comptes frauduleux certifiés.

Par ailleurs, les résultats du K-Means restent sensibles au choix du nombre de clusters et nécessitent une interprétation humaine.

Pour aller plus loin, il serait intéressant :

* d'utiliser un dataset contenant des labels réels ;
* d'ajouter des informations temporelles sur l'activité des utilisateurs ;
* d'étudier les relations entre comptes afin d'intégrer la structure du réseau social dans l'analyse.


## 4.9 Conclusion

Les deux approches se sont révélées complémentaires.

Le K-Means nous a permis de mieux comprendre la structure de la population étudiée et d'identifier plusieurs comportements atypiques sans connaissance préalable des classes.

Le Random Forest a ensuite permis d'automatiser la détection des profils atypiques selon les critères retenus et a obtenu de très bonnes performances.

Dans le cadre de ce projet, l'approche supervisée apparaît comme la plus efficace pour détecter automatiquement les profils atypiques, tandis que l'approche non supervisée constitue un excellent outil d'exploration et de compréhension des données.
