# Étape 4 : Expérimentation et analyse comparative

## 4.1 Objectif de la comparaison

Après avoir mis en œuvre une approche non supervisée (K-Means) et une approche supervisée (Random Forest), l'objectif est de comparer les résultats obtenus afin d'évaluer leur capacité à identifier des profils atypiques parmi les utilisateurs Twitter.

Dans cette étude, un profil est considéré comme atypique lorsqu'il présente un comportement significativement différent de celui observé chez la majorité des utilisateurs. Cette atypicité peut se traduire par une activité très élevée, une faible visibilité malgré une forte activité, une utilisation intensive de hashtags ou encore des comportements proches de ceux observés chez certains comptes automatisés. Cette définition repose uniquement sur les indicateurs présents dans le dataset et ne permet pas d'affirmer qu'un compte est réellement un bot ou un spammeur.


## 4.2 Hypothèses de travail

Avant d'analyser les résultats, trois hypothèses ont été formulées.

### H1 : Les profils atypiques se regroupent naturellement dans certains clusters.

Des utilisateurs ayant des comportements similaires devraient être regroupés par l'algorithme K-Means, ce qui permettrait de faire apparaître différents profils d'utilisateurs.

### H2 : Les variables liées à la popularité et à l'activité jouent un rôle important dans l'identification des profils atypiques.

Nous supposons que des variables telles que le nombre de followers, le nombre d'abonnements ou encore l'activité globale du compte constituent des indicateurs pertinents pour distinguer les profils atypiques.

### H3 : Une approche supervisée permet une détection plus précise des profils atypiques.

Une fois une définition métier du profil atypique établie, un modèle supervisé devrait être capable de détecter automatiquement ces profils avec de meilleures performances qu'une approche de clustering.


## 4.3 Résultats obtenus avec K-Means

L'algorithme K-Means a permis de segmenter automatiquement les utilisateurs Twitter en **six clusters** correspondant à des comportements distincts.

Nous avons notamment identifié :

* des créateurs de contenu engagés ;
* des comptes récents ou peu actifs ;
* plusieurs profils atypiques, notamment des spammeurs, des super-diffuseurs et des comptes très populaires assimilables à des célébrités ;
* des amplificateurs de masse relayant fortement le contenu d'autres utilisateurs.

Cette segmentation met en évidence l'hétérogénéité des comportements présents dans le dataset et permet de découvrir naturellement des groupes d'utilisateurs sans disposer de labels préalables.

Le principal avantage de K-Means est donc son caractère exploratoire : il permet d'observer la structure des données et de faire émerger différents profils comportementaux.


## 4.4 Résultats obtenus avec Random Forest

Pour l'approche supervisée, nous avons construit une variable cible à partir de plusieurs critères comportementaux :

* ratio followers/friends très faible ;
* activité anormalement élevée ;
* utilisation importante de hashtags ;
* faible visibilité.

Un utilisateur satisfaisant au moins deux de ces critères est considéré comme atypique.

Cette définition conduit à identifier **206 024 profils atypiques** parmi les **1 843 439 utilisateurs** du dataset, soit environ **11,18 %** de la population étudiée.

Les performances obtenues sont les suivantes :

* Accuracy : **97 %**
* Precision (classe atypique) : **80 %**
* Recall (classe atypique) : **99 %**
* F1-score : **88 %**

Ces résultats montrent que le modèle retrouve presque tous les profils atypiques définis par nos critères tout en maintenant un bon équilibre entre précision et rappel.


## 4.5 Analyse des variables importantes

L'analyse de l'importance des variables montre que les attributs les plus influents sont :

1. followers_count ;
2. listed_count ;
3. statuses_count ;
4. friends_count.

Le nombre de followers représente à lui seul près de **64 %** de l'importance totale du modèle.

Cela montre que la popularité et la visibilité d'un compte constituent les critères les plus déterminants pour distinguer les profils atypiques des profils normaux.

À l'inverse, certaines variables comme le nombre de réponses ou de citations ont eu une influence beaucoup plus faible dans la prise de décision du modèle.


## 4.6 Comparaison des deux approches

Les deux approches poursuivent des objectifs complémentaires.

Le **K-Means** est une méthode non supervisée dont le rôle principal est d'explorer les données. Sans connaître les classes à l'avance, il regroupe automatiquement les utilisateurs présentant des comportements similaires et permet ainsi de mettre en évidence différents profils comportementaux.

Le **Random Forest**, au contraire, est une méthode supervisée. Il apprend à partir d'une définition métier du profil atypique afin de classer automatiquement les nouveaux utilisateurs en profils normaux ou atypiques.

Ainsi, le K-Means apporte une meilleure compréhension de la structure globale des données, tandis que le Random Forest fournit un outil de classification performant capable d'automatiser la détection des profils atypiques.


## 4.7 Validation des hypothèses

### Hypothèse H1

Cette hypothèse est validée.

Le K-Means a permis d'identifier plusieurs groupes présentant des comportements nettement différents, notamment les spammeurs, les super-diffuseurs, les amplificateurs de masse et les comptes très populaires.

### Hypothèse H2

Cette hypothèse est également validée.

Les variables les plus importantes du Random Forest sont directement liées à la popularité et à l'activité des utilisateurs, en particulier le nombre de followers, le nombre d'apparitions dans des listes, le nombre total de publications et le nombre d'abonnements.

### Hypothèse H3

Les résultats obtenus confirment cette hypothèse.

Avec une accuracy de **97 %**, un rappel de **99 %** et un F1-score de **88 %**, le Random Forest est capable de détecter efficacement les profils atypiques définis dans notre étude.


## 4.8 Limites et pistes d'amélioration

La principale limite de ce travail réside dans l'absence de labels réels permettant d'identifier avec certitude les bots ou les spammeurs.

Notre variable cible repose sur une définition métier construite à partir des caractéristiques disponibles dans le dataset. Les profils détectés doivent donc être considérés comme potentiellement atypiques et non comme des comptes frauduleux certifiés.

Par ailleurs, les résultats obtenus avec K-Means restent sensibles au choix du nombre de clusters et nécessitent une interprétation humaine.

Pour prolonger cette étude, plusieurs pistes d'amélioration peuvent être envisagées :

* utiliser un dataset disposant de labels réels ;
* exploiter les relations entre comptes afin d'intégrer la structure du réseau social dans l'analyse.


## 4.9 Conclusion

Les deux approches se sont révélées complémentaires.

Le K-Means nous a permis d'explorer les données et d'identifier plusieurs profils comportementaux distincts sans connaissance préalable des classes.

Le Random Forest a ensuite permis d'automatiser la détection des profils atypiques à partir de critères métier, avec une accuracy de **97 %**, un rappel de **99 %** et un F1-score de **88 %**.

Dans le cadre de ce projet, l'approche supervisée apparaît comme la plus performante pour détecter automatiquement les profils atypiques, tandis que l'approche non supervisée constitue un excellent outil d'exploration et de compréhension des comportements des utilisateurs.
